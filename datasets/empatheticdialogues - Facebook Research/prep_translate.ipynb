# DRAFT : Pour la traduction

```python
import pandas as pd
import numpy as np
from tqdm import tqdm

PATH = "./output/"
FILENAME = "EmpatheticDialogues.tsv"

df = pd.read_csv(PATH+FILENAME, sep="\t", usecols=["Context", "Question", "Reponse"])
df.head(41)
```

```python
df.Context.unique()
```

```python
file = open("./output/a traduire2.txt", "w")
for i in tqdm(np.arange(len(df))):
    tmp = df.iloc[i]
    file.write(tmp.Question+ " \ " +tmp.Reponse + "\n")
```

```python
file = open("./output/a traduire2.txt", "r")
all_lines = file.readlines()

cpt = 0
idx = 1
limit_nbr_car = 95000

f = open("./output/traduire_0"+str(idx)+".txt", "w")
for line in all_lines:
    if (cpt + len(line) >= limit_nbr_car):
        cpt = 0
        idx+=1
        f = open("./output/traduire_0"+str(idx)+".txt", "w")
    f.write(line)
    cpt+=len(line)
```

```python
lis = np.arange(1,72)
compteur_ligne = 0
for idx in tqdm(lis) :
    f = open("./output/traduire_0"+str(idx)+".txt", "r")
    compteur_ligne += len(f.readlines())

print(compteur_ligne, len(df), len(all_lines))
```
